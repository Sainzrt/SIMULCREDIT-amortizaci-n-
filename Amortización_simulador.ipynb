```python
# --- üè¶ SIMULADOR DE CR√âDITO INTERACTIVO - By Betsy & ChatGPT ---

# Librer√≠as necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, HTML
import ipywidgets as widgets
from IPython.display import clear_output

# --- CONFIGURACI√ìN VISUAL ---
# CSS para el dise√±o fancy
style = """
<style>
body {
  background-color: #001f3f;
  color: white;
  font-family: 'Poppins', sans-serif;
}
h1 {
  text-align: center;
  color: #00bfff;
}
table {
  border-collapse: collapse;
  width: 100%;
  margin-top: 15px;
  border: 1px solid #00bfff;
}
th, td {
  border: 1px solid #00bfff;
  padding: 8px;
  text-align: center;
}
th {
  background-color: #004080;
  color: white;
}
tr:nth-child(even) {
  background-color: #002b5c;
}
.output_area pre {
  color: white;
}
</style>
"""

display(HTML(style))

# --- FUNCI√ìN PARA GENERAR LA TABLA DE AMORTIZACI√ìN ---
def generar_tabla_amortizacion(monto, tasa_anual, plazo_meses):
    tasa_mensual = tasa_anual / 12 / 100
    cuota = monto * (tasa_mensual * (1 + tasa_mensual)**plazo_meses) / ((1 + tasa_mensual)**plazo_meses - 1)

    saldo = monto
    amortizacion = []
    for i in range(1, plazo_meses + 1):
        interes = saldo * tasa_mensual
        capital = cuota - interes
        saldo -= capital
        amortizacion.append([i, round(cuota, 2), round(capital, 2), round(interes, 2), round(abs(saldo), 2)])

    df = pd.DataFrame(amortizacion, columns=["Mes", "Pago Total", "Capital", "Inter√©s", "Saldo Restante"])
    return df

# --- FUNCI√ìN DE SIMULACI√ìN INTERACTIVA ---
def simulador_credito(monto, tasa, plazo):
    clear_output(wait=True)
    display(HTML(style))
    display(HTML("<h1>üè¶ Simulador de Cr√©dito - Tabla de Amortizaci√≥n</h1>"))

    tabla = generar_tabla_amortizacion(monto, tasa, plazo)

    display(tabla.head(12))  # Muestra los primeros 12 meses (1 a√±o)
    print("\nResumen:")
    print(f"Monto del pr√©stamo: ${monto:,.2f} USD")
    print(f"Tasa de inter√©s anual: {tasa}%")
    print(f"Plazo: {plazo} meses")
    print(f"Pago mensual aproximado: ${tabla['Pago Total'][0]:,.2f}")
    print(f"Total pagado: ${tabla['Pago Total'].sum():,.2f}")
    print(f"Inter√©s total: ${tabla['Inter√©s'].sum():,.2f}")

    # --- Gr√°fico ---
    plt.figure(figsize=(8,5))
    plt.plot(tabla["Mes"], tabla["Capital"], label="Capital", linewidth=2)
    plt.plot(tabla["Mes"], tabla["Inter√©s"], label="Inter√©s", linewidth=2)
    plt.title("Distribuci√≥n de Pagos (Capital vs Inter√©s)", color='white', fontsize=14)
    plt.xlabel("Mes", color='white')
    plt.ylabel("Monto ($)", color='white')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.gca().set_facecolor("#002b5c")
    plt.gcf().patch.set_facecolor("#001f3f")
    plt.show()

# --- WIDGETS INTERACTIVOS ---
monto_slider = widgets.FloatSlider(
    value=10000, min=1000, max=100000, step=500,
    description='Monto (USD):', style={'description_width': 'initial'}, continuous_update=False
)

tasa_slider = widgets.FloatSlider(
    value=8, min=1, max=25, step=0.5,
    description='Tasa Anual (%):', style={'description_width': 'initial'}, continuous_update=False
)

plazo_slider = widgets.IntSlider(
    value=24, min=6, max=120, step=6,
    description='Plazo (meses):', style={'description_width': 'initial'}, continuous_update=False
)

ui = widgets.VBox([monto_slider, tasa_slider, plazo_slider])
out = widgets.interactive_output(simulador_credito, {'monto': monto_slider, 'tasa': tasa_slider, 'plazo': plazo_slider})

display(HTML("<h1>üí∞ Simulador de Cr√©dito Interactivo</h1>"))
display(ui, out)
```

```python
from IPython.display import display, HTML, clear_output
import pandas as pd
import io
from google.colab import output

# --- Mantengo exactamente tu mismo estilo visual y HTML (sin cambios en apariencia) ---
style = """
<style>
body {
  background-color: #f8fbff; /* Fondo blanco azulado */
  color: #0a2540; /* Azul navy profesional */
  font-family: 'Poppins', sans-serif;
}

.simucalc-container {
  background-color: white;
  border-radius: 18px;
  box-shadow: 0 10px 30px rgba(0, 64, 128, 0.1);
  width: 520px;
  margin: 70px auto;
  padding: 45px 50px;
  text-align: center;
  border: 1px solid #e1ebf5;
}

h1 {
  color: #12355b;
  font-weight: 700;
  font-size: 2.1rem;
  letter-spacing: 1px;
  margin-bottom: 35px;
}

label {
  display: block;
  font-size: 0.95rem;
  margin: 18px 0 6px;
  color: #1c3d6d;
  text-align: left;
  font-weight: 500;
}

input {
  width: 100%;
  padding: 12px;
  border-radius: 10px;
  border: 1px solid #a7c4e0;
  outline: none;
  font-size: 1rem;
  background-color: #f3f8fc;
  color: #0a2540;
  transition: 0.3s;
}

input:focus {
  border-color: #3075b5;
  background-color: #eef6ff;
  box-shadow: 0 0 6px rgba(0, 102, 204, 0.2);
}

input::placeholder {
  color: #93adc7;
}

button {
  background-color: #164b88;
  border: none;
  color: white;
  padding: 13px 28px;
  border-radius: 10px;
  font-size: 1rem;
  cursor: pointer;
  margin-top: 30px;
  font-weight: 600;
  letter-spacing: 0.3px;
  transition: 0.3s ease-in-out;
  width: 100%;
}

button:hover {
  background-color: #1f5aa8;
  transform: translateY(-2px);
  box-shadow: 0 4px 15px rgba(22, 75, 136, 0.3);
}

footer {
  text-align: center;
  color: #5d7da1;
  margin-top: 30px;
  font-size: 0.9rem;
  letter-spacing: 0.4px;
}

.brand-line {
  width: 70px;
  height: 3px;
  background-color: #164b88;
  margin: -15px auto 30px;
  border-radius: 3px;
}

#resultado {
  width: 520px;            /* coincide con el ancho del container para que quede centrado */
  margin: 24px auto 60px;  /* centrado y separaci√≥n */
  background: white;
  padding: 22px;
  border-radius: 12px;
  box-shadow: 0 5px 20px rgba(0,0,0,0.03);
  border: 1px solid #e1ebf5;
  text-align: left;
}

.table-styles {
  width: 100%;
  border-collapse: collapse;
  margin-top: 10px;
  font-size: 0.95rem;
  color: #0a2540;
}

.table-styles th, .table-styles td {
  padding: 10px 8px;
  border-bottom: 1px solid #eef4fb;
  text-align: right;
}

.table-styles th {
  text-align: left;
  color: #12355b;
  font-weight: 600;
  background: transparent;
}
.download-btn {
  display:inline-block;
  margin-top:12px;
  background-color:#164b88;
  color:white;
  padding:10px 16px;
  border-radius:8px;
  text-decoration:none;
  font-weight:600;
}
</style>
"""

# HTML id√©ntico pero el onclick ahora Llama DIRECTAMENTE al callback backend ('calc_credit_backend')
html = """
<div class="simucalc-container">
  <h1>SIMULCREDIT</h1>
  <div class="brand-line"></div>

  <label for="monto">Monto del cr√©dito (USD):</label>
  <input type="number" id="monto" placeholder="Ej. 15000" />

  <label for="tasa">Tasa de inter√©s anual (%):</label>
  <input type="number" id="tasa" placeholder="Ej. 7.5" />

  <label for="plazo">Plazo (meses):</label>
  <input type="number" id="plazo" placeholder="Ej. 36" />

  <!-- Aqu√≠ est√° el cambio clave: invocamos calc_credit_backend directamente con los 3 valores -->
  <button onclick="google.colab.kernel.invokeFunction('calc_credit_backend', [parseFloat(document.getElementById('monto').value) || 0, parseFloat(document.getElementById('tasa').value) || 0, parseInt(document.getElementById('plazo').value) || 0], {})">Calcular cr√©dito</button>
</div>

<div id="resultado"></div>

<footer>¬© 2025 Simulcredit | Plataforma de simulaci√≥n financiera</footer>
"""

display(HTML(style + html))


# --- Funci√≥n de c√°lculo (backend) ---
def generar_tabla_amortizacion(monto, tasa_anual, plazo_meses):
    # si tasa es cero o negativa, manejamos caso simple (cuota = monto/plazo)
    if tasa_anual <= 0:
        cuota = monto / plazo_meses
        saldo = monto
        datos = []
        for i in range(1, plazo_meses + 1):
            interes = 0.0
            capital = cuota
            saldo -= capital
            datos.append([i, round(cuota,2), round(capital,2), round(interes,2), round(abs(saldo),2)])
        df = pd.DataFrame(datos, columns=["Mes", "Pago total", "Capital", "Inter√©s", "Saldo restante"])
        return df

    tasa_mensual = tasa_anual / 12 / 100
    # f√≥rmula de cuota (anual->mensual)
    cuota = monto * (tasa_mensual * (1 + tasa_mensual)**plazo_meses) / ((1 + tasa_mensual)**plazo_meses - 1)
    saldo = monto
    datos = []
    for i in range(1, plazo_meses + 1):
        interes = saldo * tasa_mensual
        capital = cuota - interes
        saldo -= capital
        datos.append([i, round(cuota, 2), round(capital, 2), round(interes, 2), round(abs(saldo), 2)])
    df = pd.DataFrame(datos, columns=["Mes", "Pago total", "Capital", "Inter√©s", "Saldo restante"])
    return df

def calc_credit_backend(monto, tasa, plazo):
    # Esta funci√≥n ser√° llamada desde el navegador (JS) con los 3 valores
    clear_output(wait=True)          # limpiamos para reimprimir la interfaz y el resultado debajo conservando el mismo look
    display(HTML(style + html))      # volvemos a mostrar exactamente la interfaz (id√©ntica)

    # validaciones simples
    try:
        monto = float(monto)
        tasa = float(tasa)
        plazo = int(plazo)
    except:
        display(HTML("<div id='resultado'><b style='color:#b30000;'>Por favor ingresa valores num√©ricos v√°lidos.</b></div>"))
        return

    if monto <= 0 or plazo <= 0 or tasa < 0:
        display(HTML("<div id='resultado'><b style='color:#b30000;'>Por favor ingresa valores v√°lidos (monto>0, plazo>0, tasa>=0).</b></div>"))
        return

    # generar tabla
    df = generar_tabla_amortizacion(monto, tasa, plazo)

    # Guardar Excel localmente
    excel_buffer = io.BytesIO()
    df.to_excel(excel_buffer, index=False, engine='openpyxl')
    excel_buffer.seek(0)
    with open("Tabla_Amortizacion_Simulcredit.xlsx", "wb") as f:
        f.write(excel_buffer.read())

    # Construir HTML de la tabla con estilos (alineaci√≥n y formato)
    # Formateamos n√∫meros con separadores y 2 decimales
    def fmt(x):
        return f"{x:,.2f}"

    # cabecera del bloque resultado (manteniendo ancho y centrado)
    html_tabla = """
    <div id="resultado">
      <h3 style="color:#12355b; margin-top:0;">Tabla de Amortizaci√≥n</h3>
      <table class="table-styles">
        <thead>
          <tr>
            <th>Mes</th>
            <th>Pago total (USD)</th>
            <th>Capital (USD)</th>
            <th>Inter√©s (USD)</th>
            <th>Saldo restante (USD)</th>
          </tr>
        </thead>
        <tbody>
    """
    for _, row in df.iterrows():
        html_tabla += f"""
          <tr>
            <td style="text-align:left;">{int(row['Mes'])}</td>
            <td>{fmt(row['Pago total'])}</td>
            <td>{fmt(row['Capital'])}</td>
            <td>{fmt(row['Inter√©s'])}</td>
            <td>{fmt(row['Saldo restante'])}</td>
          </tr>
        """
    html_tabla += """
        </tbody>
      </table>
      <a class="download-btn" href="Tabla_Amortizacion_Simulcredit.xlsx" download>Descargar en Excel</a>
    </div>
    """

    display(HTML(html_tabla))


# Registramos la funci√≥n para que JS pueda invocarla
output.register_callback('calc_credit_backend', calc_credit_backend)
```

```python
# Pegar y ejecutar en Google Colab
from IPython.display import display, HTML, Javascript
import pandas as pd
import io, base64
import matplotlib.pyplot as plt
import numpy as np
from google.colab import output, files

# ----------------- ESTILO (ampliado para animaciones y paneles) -----------------
style = """
<style>
/* Base (tu estilo original, ligeramente ampliado) */
body {
  background-color: #f8fbff;
  color: #0a2540;
  font-family: 'Poppins', sans-serif;
}
.simucalc-container {
  background-color: white;
  border-radius: 18px;
  box-shadow: 0 10px 30px rgba(0, 64, 128, 0.1);
  width: 520px;
  margin: 40px auto;
  padding: 30px 36px;
  text-align: center;
  border: 1px solid #e1ebf5;
}
h1 { color:#12355b; font-weight:700; font-size:2.1rem; margin-bottom:18px;}
label{ display:block; font-size:0.95rem; margin:12px 0 6px; color:#1c3d6d; text-align:left; font-weight:500;}
input{ width:100%; padding:12px; border-radius:10px; border:1px solid #a7c4e0; outline:none; font-size:1rem; background:#f3f8fc;}
input:focus{ border-color:#3075b5; background:#eef6ff; box-shadow:0 0 6px rgba(0,102,204,0.12);}
button{ background-color:#164b88; border:none; color:white; padding:12px 18px; border-radius:10px; font-size:1rem; cursor:pointer; margin-top:18px; font-weight:600; width:100%}
button.small{ padding:8px 12px; font-size:0.95rem; width:auto; margin:6px;}
button:hover{ background:#1f5aa8; transform:translateY(-2px); box-shadow:0 6px 18px rgba(22,75,136,0.2);}

/* Resultado / panels */
#resultado { width: 520px; margin: 18px auto 60px; background: white; padding: 18px; border-radius: 12px; box-shadow: 0 5px 20px rgba(0,0,0,0.03); border:1px solid #e1ebf5; text-align:left;}
.table-styles { width:100%; border-collapse:collapse; margin-top:10px; font-size:0.95rem; color:#0a2540;}
.table-styles th, .table-styles td { padding:10px 8px; border-bottom:1px solid #eef4fb; text-align:right;}
.table-styles th{ text-align:left; color:#12355b; font-weight:600; background:transparent;}

/* download & option buttons */
.download-btn { display:inline-block; margin-top:10px; background-color:#164b88; color:white; padding:10px 14px; border-radius:8px; text-decoration:none; font-weight:600; cursor:pointer;}
.action-row { display:flex; gap:10px; flex-wrap:wrap; margin-top:12px; }

/* Cute/animated educational panel */
.educ-panel {
  padding:16px; border-radius:12px; background: linear-gradient(135deg,#fff6fb,#f0fbff);
  border: 1px dashed #f6e6ff; box-shadow: 0 8px 30px rgba(0,0,0,0.03); animation: floaty 6s ease-in-out infinite;
}
@keyframes floaty {
  0% { transform: translateY(0px); }
  50% { transform: translateY(-6px); }
  100% { transform: translateY(0px); }
}
.card {
  background: white; border-radius:10px; padding:12px; margin-bottom:10px; box-shadow:0 5px 18px rgba(17,44,90,0.03);
}
.icon { font-size:28px; margin-right:10px; vertical-align:middle; }

/* small responsive */
@media (max-width:640px) {
  .simucalc-container, #resultado { width:92%; margin-left:4%; margin-right:4%;}
}
</style>
"""

# ----------------- HTML PRINCIPAL (agrega los botones nuevos) -----------------
html = """
<div class="simucalc-container">
  <h1>SIMULCREDIT</h1>
  <div class="brand-line" style="width:70px;height:3px;background:#164b88;margin:-8px auto 12px;border-radius:3px;"></div>

  <label for="monto">Monto del cr√©dito (USD):</label>
  <input type="number" id="monto" placeholder="Ej. 15000" />

  <label for="tasa">Tasa de inter√©s anual (%):</label>
  <input type="number" id="tasa" placeholder="Ej. 7.5" />

  <label for="plazo">Plazo (meses):</label>
  <input type="number" id="plazo" placeholder="Ej. 36" />

  <button onclick="google.colab.kernel.invokeFunction('calc_credit_backend',
    [parseFloat(document.getElementById('monto').value) || 0,
     parseFloat(document.getElementById('tasa').value) || 0,
     parseInt(document.getElementById('plazo').value) || 0], {})">
     Calcular cr√©dito
  </button>

  <div style="margin-top:10px;" class="action-row">
    <button class="download-btn" onclick="google.colab.kernel.invokeFunction('finanbot', [], {})">üí¨ Finanbot</button>
    <button class="download-btn" onclick="google.colab.kernel.invokeFunction('mostrar_dashboard', [], {})">üìä Dashboard</button>
    <button class="download-btn" onclick="google.colab.kernel.invokeFunction('mostrar_simulador_form', [], {})">üîÅ Simulador de escenarios</button>
  </div>

  <div style="margin-top:12px;" class="action-row">
    <button class="download-btn" onclick="google.colab.kernel.invokeFunction('componente_educativo', [], {})">üìö Componente educativo</button>
    <button class="download-btn" onclick="google.colab.kernel.invokeFunction('integrar_tasas_menu', [], {})">üè¶ Integrar tasas bancarias internacionales</button>
  </div>
</div>

<div id="resultado"></div>
<footer style="text-align:center; color:#5d7da1; margin-top:18px; font-size:0.9rem;">¬© 2025 Simulcredit | Plataforma de simulaci√≥n financiera</footer>
"""

display(HTML(style + html))

# ----------------- FUNCIONES DE C√ÅLCULO (base s√≥lida) -----------------
def generar_tabla_amortizacion(monto, tasa_anual, plazo_meses):
    # monto: float, tasa_anual: porcentaje (e.g. 7.5), plazo_meses: int
    if plazo_meses <= 0:
        return pd.DataFrame([], columns=["Mes", "Pago total", "Capital", "Inter√©s", "Saldo restante"])
    if tasa_anual <= 0:
        cuota = monto / plazo_meses
        saldo = monto
        datos = []
        for i in range(1, plazo_meses + 1):
            interes = 0.0
            capital = cuota
            saldo -= capital
            datos.append([i, round(cuota,2), round(capital,2), round(interes,2), round(max(saldo,0),2)])
        return pd.DataFrame(datos, columns=["Mes", "Pago total", "Capital", "Inter√©s", "Saldo restante"])

    tasa_mensual = tasa_anual / 12 / 100
    cuota = monto * (tasa_mensual * (1 + tasa_mensual)**plazo_meses) / ((1 + tasa_mensual)**plazo_meses - 1)
    saldo = monto
    datos = []
    for i in range(1, plazo_meses + 1):
        interes = saldo * tasa_mensual
        capital = cuota - interes
        saldo -= capital
        datos.append([i, round(cuota, 2), round(capital, 2), round(interes, 2), round(max(saldo,0), 2)])
    return pd.DataFrame(datos, columns=["Mes", "Pago total", "Capital", "Inter√©s", "Saldo restante"])

# Variables globales para mantener estado
ultima_tasa = None
ultimo_plazo = None
ultimo_monto = None
ultima_df = None

# ----------------- CALLBACK PRINCIPAL: calcular cr√©dito -----------------
def calc_credit_backend(monto, tasa, plazo):
    global ultima_tasa, ultimo_plazo, ultimo_monto, ultima_df
    try:
        monto = float(monto)
        tasa = float(tasa)
        plazo = int(plazo)
    except:
        display(Javascript("""document.getElementById('resultado').innerHTML =
            '<b style="color:#b30000;">Por favor ingresa valores v√°lidos.</b>'"""))
        return
    if monto <= 0 or plazo <= 0 or tasa < 0:
        display(Javascript("""document.getElementById('resultado').innerHTML =
            '<b style="color:#b30000;'>Por favor ingresa valores v√°lidos (monto>0, plazo>0, tasa‚â•0).</b>'"""))
        return

    df = generar_tabla_amortizacion(monto, tasa, plazo)
    ultima_df = df.copy()
    ultima_tasa, ultimo_plazo, ultimo_monto = tasa, plazo, monto

    # Guardar Excel
    nombre_archivo = "Tabla_Amortizacion_Simulcredit.xlsx"
    excel_buffer = io.BytesIO()
    df.to_excel(excel_buffer, index=False, engine='openpyxl')
    excel_buffer.seek(0)
    with open(nombre_archivo, "wb") as f:
        f.write(excel_buffer.read())

    # Resumen crediticio (resumen inteligente)
    cuota = df['Pago total'].iloc[0] if not df.empty else 0
    total_pagado = df['Pago total'].sum()
    total_intereses = df['Inter√©s'].sum()
    porcentaje_intereses = (total_intereses / total_pagado * 100) if total_pagado>0 else 0

    resumen_html = f"""
      <div style="background:#eef6ff; padding:14px; border-radius:10px; margin-bottom:12px;">
        <b>üìã Resumen crediticio</b><br>
        <b>Cuota mensual:</b> ${cuota:,.2f} &nbsp; ‚Ä¢ &nbsp;
        <b>Total pagado:</b> ${total_pagado:,.2f} &nbsp; ‚Ä¢ &nbsp;
        <b>Intereses totales:</b> ${total_intereses:,.2f} <br>
        <small>Los intereses representan <b>{porcentaje_intereses:.2f}%</b> del total pagado.</small>
      </div>
    """

    # Tabla en HTML
    def fmt(x): return f"{x:,.2f}"
    html_tabla = """
    <h3 style="color:#12355b; margin-top:0;">Tabla de Amortizaci√≥n</h3>
    <table class="table-styles">
      <thead>
        <tr>
          <th>Mes</th><th>Pago total (USD)</th><th>Capital (USD)</th>
          <th>Inter√©s (USD)</th><th>Saldo restante (USD)</th>
        </tr>
      </thead><tbody>
    """
    for _, row in df.iterrows():
        html_tabla += f"""
          <tr>
            <td style='text-align:left;'>{int(row['Mes'])}</td>
            <td>{fmt(row['Pago total'])}</td>
            <td>{fmt(row['Capital'])}</td>
            <td>{fmt(row['Inter√©s'])}</td>
            <td>{fmt(row['Saldo restante'])}</td>
          </tr>"""
    html_tabla += "</tbody></table>"

    # Botones: descargar, Finanbot (separado), simulador y dashboard ya en UI
    botones = """
      <div style="margin-top:12px;">
        <button class='download-btn' onclick="google.colab.kernel.invokeFunction('descargar_excel', [], {})">üì• Descargar Excel</button>
        <button class='download-btn' style="margin-left:8px;" onclick="google.colab.kernel.invokeFunction('finanbot', [], {})">üí¨ Finanbot</button>
      </div>
    """

    display(Javascript(f"""
      document.getElementById('resultado').innerHTML = `{resumen_html + html_tabla + botones}`;
      window.scrollTo(0, document.getElementById('resultado').offsetTop - 30);
    """))

# ----------------- DESCARGAR EXCEL -----------------
def descargar_excel():
    try:
        files.download("Tabla_Amortizacion_Simulcredit.xlsx")
    except Exception as e:
        display(Javascript(f"document.getElementById('resultado').innerHTML += '<div style=\"color:#b30000;\">Error al descargar: {str(e)}</div>';"))

# ----------------- FINANBOT (recomendaci√≥n personalizada) -----------------
def finanbot():
    global ultima_tasa, ultimo_plazo, ultimo_monto, ultima_df
    if ultima_monto is None:
        display(Javascript("""document.getElementById('resultado').innerHTML +=
            '<div style="margin-top:14px;color:#b30000"><b>Primero calcula un cr√©dito para que Finanbot pueda darte recomendaciones.</b></div>'"""))
        return

    tasa = ultima_tasa
    plazo = ultimo_plazo
    monto = ultima_monto
    df = ultima_df if ultima_df is not None else generar_tabla_amortizacion(monto, tasa, plazo)

    # Recomendaci√≥n de plazo
    if plazo <= 12:
        rec_plazo = "Si tu flujo lo permite, extender a 24 meses puede reducir presi√≥n mensual; sin embargo, eval√∫a el costo total."
    elif plazo <= 36:
        rec_plazo = "Rango equilibrado: cuotas manejables y costo total moderado."
    else:
        rec_plazo = "Si puedes, acorta plazo para reducir intereses totales."

    # Recomendaci√≥n de tasa: sugerir negociar debajo de 85% de la tasa actual (heur√≠stica)
    tasa_objetivo = tasa * 0.85
    rec_tasa = f"Busca negociar una tasa menor a {tasa_objetivo:.2f}% anual. Cada punto porcentual puede ahorrar bastante en intereses."

    # Consejo pr√°ctico de pago extra
    pago_extra_recomendado = round(monto * 0.05, 2)  # ejemplo 5% del monto como pago extra
    rec_extra = f"Si puedes pagar un extra mensual aproximado de ${pago_extra_recomendado:,.2f}, reducir√°s el plazo e intereses significativamente."

    # Mostrar
    content = f"""
      <div style="margin-top:16px; background:#f1f7ff; padding:14px; border-radius:10px; border:1px solid #d7e7ff;">
        <h3 style="color:#12355b;">ü§ñ Finanbot ‚Äî tu consejero financiero</h3>
        <p style="line-height:1.5;">üí° <b>Plazo recomendado:</b> {rec_plazo}</p>
        <p style="line-height:1.5;">üìâ <b>Tasa ideal:</b> {rec_tasa}</p>
        <p style="line-height:1.5;">üîÅ <b>Pago extra sugerido:</b> {rec_extra}</p>
        <small style="color:#4b6b8d;">*Recomendaciones heur√≠sticas. Personaliza seg√∫n tus ingresos y objetivos.</small>
      </div>
    """
    display(Javascript(f"document.getElementById('resultado').innerHTML += `{content}`;"))

# ----------------- SIMULADOR DE ESCENARIOS: mostrar formulario -----------------
def mostrar_simulador_form():
    # Insertar en resultado un peque√±o formulario para comparar escenario alternativo
    js = """
    const base = document.getElementById('resultado');
    base.innerHTML += `
      <div style="margin-top:14px; padding:12px; border-radius:10px; background:#fff8ee; border:1px solid #fff0d6;" id="simulador-panel">
        <h3 style="margin-top:0; color:#12355b;">üîÅ Simulador de escenarios</h3>
        <div style="display:flex; gap:8px; flex-wrap:wrap;">
          <div style="flex:1;">
            <label style="font-size:0.92rem; color:#1c3d6d;">Tasa alternativa (%)</label>
            <input id="sim_tasa" type="number" placeholder="Ej. 6.5" style="padding:8px; border-radius:8px; border:1px solid #cfe4ff; width:100%;" />
          </div>
          <div style="width:120px;">
            <label style="font-size:0.92rem; color:#1c3d6d;">Plazo (meses)</label>
            <input id="sim_plazo" type="number" placeholder="Ej. 24" style="padding:8px; border-radius:8px; border:1px solid #cfe4ff; width:100%;" />
          </div>
        </div>
        <div style="margin-top:10px;">
          <button class="download-btn" onclick="google.colab.kernel.invokeFunction('procesar_simulador', [
            parseFloat(document.getElementById('sim_tasa').value) || 0,
            parseInt(document.getElementById('sim_plazo').value) || 0
          ], {})">Comparar escenarios</button>
        </div>
      </div>
    `;
    """
    display(Javascript(js))

def procesar_simulador(tasa_alt, plazo_alt):
    # Compara √∫ltimo cr√©dito con alternativa dada
    global ultima_tasa, ultimo_plazo, ultimo_monto, ultima_df
    if ultima_monto is None:
        display(Javascript("""document.getElementById('resultado').innerHTML +=
            '<div style="color:#b30000; margin-top:10px;"><b>Primero calcula un cr√©dito para usar el simulador.</b></div>'"""))
        return
    if tasa_alt <= 0 or plazo_alt <= 0:
        display(Javascript("""document.getElementById('resultado').innerHTML +=
            '<div style="color:#b30000; margin-top:10px;">Ingresa tasa alternativa y plazo v√°lidos.</div>'"""))
        return

    monto = ultimo_monto
    df_base = ultima_df if ultima_df is not None else generar_tabla_amortizacion(monto, ultima_tasa, ultimo_plazo)
    df_alt = generar_tabla_amortizacion(monto, tasa_alt, plazo_alt)

    total_intereses_base = df_base['Inter√©s'].sum()
    total_intereses_alt = df_alt['Inter√©s'].sum()
    ahorro = total_intereses_base - total_intereses_alt
    mejor = "escenario alternativo" if ahorro>0 else "escenario actual (no conviene cambiar)"
    detalle_html = f"""
      <div style="margin-top:12px; padding:12px; border-radius:10px; background:#eefaf1; border:1px solid #dff3e6;">
        <h4 style="margin-top:0;color:#12355b;">Resultado de la comparaci√≥n</h4>
        <p><b>Intereses totales ‚Äî Actual:</b> ${total_intereses_base:,.2f}</p>
        <p><b>Intereses totales ‚Äî Alternativo:</b> ${total_intereses_alt:,.2f}</p>
        <p><b>Ahorro estimado al cambiar:</b> ${ahorro:,.2f}</p>
        <p><b>Recomendaci√≥n:</b> Seg√∫n este c√°lculo, te conviene: <b>{mejor}</b></p>
      </div>
    """

    # Generar mini gr√°fico comparativo (barras) y devolver como base64 PNG
    plt.figure(figsize=(4,2.2))
    bars = [total_intereses_base, total_intereses_alt]
    labels = ['Actual', 'Alternativo']
    x = np.arange(len(labels))
    plt.bar(x, bars)
    plt.xticks(x, labels)
    plt.ylabel('Intereses totales (USD)')
    plt.tight_layout()
    imgbuf = io.BytesIO()
    plt.savefig(imgbuf, format='png', bbox_inches='tight', dpi=120)
    plt.close()
    imgbuf.seek(0)
    b64 = base64.b64encode(imgbuf.read()).decode('utf-8')
    graf_html = f'<div style="margin-top:10px;"><img src="data:image/png;base64,{b64}" style="max-width:100%; border-radius:8px;"/></div>'

    display(Javascript(f"document.getElementById('resultado').innerHTML += `{detalle_html + graf_html}`;"))

# ----------------- DASHBOARD (genera gr√°ficos si existe cr√©dito) -----------------
def mostrar_dashboard():
    global ultima_df, ultimo_monto, ultima_tasa, ultimo_plazo
    if ultima_df is None:
        display(Javascript("""document.getElementById('resultado').innerHTML +=
            '<div style="color:#b30000; margin-top:10px;"><b>Calcula un cr√©dito primero para ver el dashboard.</b></div>'"""))
        return

    df = ultima_df.copy()
    # Gr√°fico 1: Saldo restante en el tiempo
    plt.figure(figsize=(5,2.2))
    plt.plot(df['Mes'], df['Saldo restante'], marker='o')
    plt.title('Saldo restante vs Mes')
    plt.xlabel('Mes')
    plt.ylabel('Saldo restante (USD)')
    plt.tight_layout()
    buf1 = io.BytesIO()
    plt.savefig(buf1, format='png', dpi=120, bbox_inches='tight')
    plt.close()
    buf1.seek(0)
    b641 = base64.b64encode(buf1.read()).decode('utf-8')

    # Gr√°fico 2: Pastel Inter√©s vs Capital total
    total_intereses = df['Inter√©s'].sum()
    total_capital = df['Capital'].sum()
    plt.figure(figsize=(3.5,3.5))
    plt.pie([total_capital, total_intereses], labels=['Capital','Inter√©s'], autopct='%1.1f%%', startangle=90)
    plt.axis('equal')
    buf2 = io.BytesIO()
    plt.savefig(buf2, format='png', dpi=120, bbox_inches='tight')
    plt.close()
    buf2.seek(0)
    b642 = base64.b64encode(buf2.read()).decode('utf-8')

    # Dashboard HTML
    dash = f"""
      <div style="margin-top:12px; padding:12px; border-radius:10px; background:#fff; border:1px solid #e8f1ff;">
        <h3 style="margin-top:0;color:#12355b;">üìä Dashboard financiero</h3>
        <div style="display:flex; gap:10px; flex-wrap:wrap;">
          <div style="flex:1; min-width:240px;">
            <b>Saldo restante (proyecci√≥n)</b>
            <div style="margin-top:8px;"><img src="data:image/png;base64,{b641}" style="width:100%; border-radius:8px;" /></div>
          </div>
          <div style="width:200px;">
            <b>Composici√≥n de pago</b>
            <div style="margin-top:8px;"><img src="data:image/png;base64,{b642}" style="width:100%; border-radius:8px;" /></div>
          </div>
        </div>
        <p style="color:#4b6b8d; margin-top:10px;"><small>Gr√°ficos generados a partir de la tabla actual.</small></p>
      </div>
    """
    display(Javascript(f"document.getElementById('resultado').innerHTML += `{dash}`;"))

# ----------------- COMPONENTE EDUCATIVO (animado y cute) -----------------
def componente_educativo():
    content = """
    <div class="educ-panel" style="margin-top:12px;">
      <h3 style="margin-top:0; color:#12355b;">üåü Componente educativo</h3>
      <div class="card">
        <div><span class="icon">üìò</span><b>¬øQu√© es la tasa de inter√©s?</b></div>
        <div style="margin-top:8px; color:#315a82;">La tasa es el costo que pagas por pedir prestado. Se expresa en porcentaje anual. Una tasa m√°s baja reduce el costo total del cr√©dito.</div>
      </div>
      <div class="card">
        <div><span class="icon">üßæ</span><b>Plazo vs cuota</b></div>
        <div style="margin-top:8px; color:#315a82;">A mayor plazo, la cuota mensual baja, pero pagas m√°s intereses en total. A menor plazo, cuotas m√°s altas pero menos intereses.</div>
      </div>
      <div class="card">
        <div><span class="icon">‚ö°</span><b>Pagos extra</b></div>
        <div style="margin-top:8px; color:#315a82;">Realizar pagos extra reduce saldo y tiempo del cr√©dito. Incluso pagos peque√±os recurrentes ayudan mucho.</div>
      </div>
      <div style="text-align:center; margin-top:8px;">
        <button class="small" onclick="document.getElementById('resultado').innerHTML = ''">Cerrar</button>
      </div>
    </div>
    """
    display(Javascript(f"document.getElementById('resultado').innerHTML = `{content}`;"))

# ----------------- INTEGRAR TASAS BANCARIAS (menu + mock data) -----------------
# Aqu√≠ usamos tasas est√°ticas de ejemplo. Puedes conectar APIs reales despu√©s.
tasas_mock = {
    "mexico": {"name":"Banco de M√©xico", "tasa_ano":"11.00"},
    "colombia": {"name":"Banco de la Rep√∫blica (Colombia)", "tasa_ano":"12.50"},
    "eeuu": {"name":"Fed (EE.UU.) - referencia", "tasa_ano":"5.25"},
    "brasil": {"name":"BCB (Brasil)", "tasa_ano":"13.75"},
    "espana": {"name":"Banco de Espa√±a (EUR ref)", "tasa_ano":"3.50"},
}

def integrar_tasas_menu():
    menu_html = """
      <div style="margin-top:12px; padding:12px; border-radius:10px; background:#fff; border:1px solid #e8f1ff;">
        <h3 style="margin-top:0;color:#12355b;">üè¶ Tasas bancarias internacionales</h3>
        <p style="color:#4b6b8d;">Selecciona un banco para ver una tasa de referencia (ejemplo).</p>
        <div style="display:flex; gap:8px; flex-wrap:wrap;">
          <button class="download-btn" onclick="google.colab.kernel.invokeFunction('mostrar_tasas', ['mexico'], {})">Banco de M√©xico</button>
          <button class="download-btn" onclick="google.colab.kernel.invokeFunction('mostrar_tasas', ['colombia'], {})">Colombia</button>
          <button class="download-btn" onclick="google.colab.kernel.invokeFunction('mostrar_tasas', ['eeuu'], {})">EE.UU.</button>
          <button class="download-btn" onclick="google.colab.kernel.invokeFunction('mostrar_tasas', ['brasil'], {})">Brasil</button>
          <button class="download-btn" onclick="google.colab.kernel.invokeFunction('mostrar_tasas', ['espana'], {})">Espa√±a</button>
        </div>
      </div>
    """
    display(Javascript(f"document.getElementById('resultado').innerHTML = `{menu_html}`;"))

def mostrar_tasas(key):
    d = tasas_mock.get(key, None)
    if d is None:
        display(Javascript("document.getElementById('resultado').innerHTML += '<div style=color:#b30000;>Banco no encontrado.</div>'"))
        return
    # Muestra desglose (mock)
    html = f"""
      <div style="margin-top:12px; padding:12px; border-radius:10px; background:#f8fff6; border:1px solid #e8f6ea;">
        <h4 style="margin-top:0;color:#12355b;">{d['name']}</h4>
        <p><b>Tasa de referencia (ejemplo):</b> {d['tasa_ano']}% anual</p>
        <p style="color:#4b6b8d;"><small>Nota: Estos valores son ejemplos est√°ticos. Para producci√≥n conecte la API oficial de cada banco o fuente (Banxico, BanRep, Fed, BCB, BCE/Banco de Espa√±a).</small></p>
      </div>
    """
    display(Javascript(f"document.getElementById('resultado').innerHTML += `{html}`;"))

# ----------------- REGISTRO DE CALLBACKS -----------------
output.register_callback('calc_credit_backend', calc_credit_backend)
output.register_callback('descargar_excel', descargar_excel)
output.register_callback('finanbot', finanbot)
output.register_callback('mostrar_simulador_form', mostrar_simulador_form)
output.register_callback('procesar_simulador', procesar_simulador)
output.register_callback('mostrar_dashboard', mostrar_dashboard)
output.register_callback('componente_educativo', componente_educativo)
output.register_callback('integrar_tasas_menu', integrar_tasas_menu)
output.register_callback('mostrar_tasas', mostrar_tasas)

print("Simulcredit listo ‚Äî usa el formulario arriba y prueba las nuevas funciones: Finanbot, Dashboard, Simulador, Componente educativo e Integrar tasas.")
```

```python
# Pega y ejecuta TODO este bloque en una sola celda en Google Colab

from IPython.display import display, HTML, Javascript
import pandas as pd
import io, re, os, time, json, base64
from google.colab import output, files
import matplotlib.pyplot as plt
import numpy as np
import requests

# ---------------------------
# === ESTILO (id√©ntico) ===
# ---------------------------
style = """
<style>
body {
  background-color: #f8fbff;
  color: #0a2540;
  font-family: 'Poppins', sans-serif;
}

.simucalc-container {
  background-color: white;
  border-radius: 18px;
  box-shadow: 0 10px 30px rgba(0, 64, 128, 0.1);
  width: 520px;
  margin: 70px auto;
  padding: 45px 50px;
  text-align: center;
  border: 1px solid #e1ebf5;
}

h1 {
  color: #12355b;
  font-weight: 700;
  font-size: 2.1rem;
  letter-spacing: 1px;
  margin-bottom: 35px;
}

label {
  display: block;
  font-size: 0.95rem;
  margin: 18px 0 6px;
  color: #1c3d6d;
  text-align: left;
  font-weight: 500;
}

input {
  width: 100%;
  padding: 12px;
  border-radius: 10px;
  border: 1px solid #a7c4e0;
  outline: none;
  font-size: 1rem;
  background-color: #f3f8fc;
  color: #0a2540;
  transition: 0.3s;
}

input:focus {
  border-color: #3075b5;
  background-color: #eef6ff;
  box-shadow: 0 0 6px rgba(0, 102, 204, 0.2);
}

button {
  background-color: #164b88;
  border: none;
  color: white;
  padding: 13px 28px;
  border-radius: 10px;
  font-size: 1rem;
  cursor: pointer;
  margin-top: 30px;
  font-weight: 600;
  letter-spacing: 0.3px;
  transition: 0.3s ease-in-out;
  width: 100%;
}

button:hover {
  background-color: #1f5aa8;
  transform: translateY(-2px);
  box-shadow: 0 4px 15px rgba(22, 75, 136, 0.3);
}

footer {
  text-align: center;
  color: #5d7da1;
  margin-top: 30px;
  font-size: 0.9rem;
  letter-spacing: 0.4px;
}

.brand-line {
  width: 70px;
  height: 3px;
  background-color: #164b88;
  margin: -15px auto 30px;
  border-radius: 3px;
}

#resultado {
  width: 520px;
  margin: 24px auto 60px;
  background: white;
  padding: 22px;
  border-radius: 12px;
  box-shadow: 0 5px 20px rgba(0,0,0,0.03);
  border: 1px solid #e1ebf5;
  text-align: left;
}

.table-styles {
  width: 100%;
  border-collapse: collapse;
  margin-top: 10px;
  font-size: 0.95rem;
  color: #0a2540;
}

.table-styles th, .table-styles td {
  padding: 10px 8px;
  border-bottom: 1px solid #eef4fb;
  text-align: right;
}

.table-styles th {
  text-align: left;
  color: #12355b;
  font-weight: 600;
  background: transparent;
}

.download-btn {
  display:inline-block;
  margin-top:15px;
  background-color:#164b88;
  color:white;
  padding:10px 18px;
  border-radius:8px;
  text-decoration:none;
  font-weight:600;
  cursor:pointer;
  transition:0.3s;
}

.download-btn:hover {
  background-color:#1f5aa8;
  transform:translateY(-2px);
}

/* Finanbot widget en esquina inferior derecha (no modifica tu formato principal) */
#finanbot_widget {
  position: fixed;
  right: 18px;
  bottom: 18px;
  width: 340px;
  max-width: 92%;
  background: linear-gradient(180deg, #ffffff, #f1f7ff);
  border-radius: 12px;
  box-shadow: 0 10px 30px rgba(0,64,128,0.12);
  border: 1px solid #e6f0ff;
  z-index: 9999;
  overflow: hidden;
  display: none; /* aparece solo en segunda p√°gina */
}

#finanbot_header {
  background: linear-gradient(90deg,#164b88,#1f5aa8);
  color:white;
  padding:10px 12px;
  font-weight:700;
  display:flex;
  justify-content:space-between;
  align-items:center;
}
#finanbot_messages { height:220px; overflow:auto; padding:10px; background:white;}
#finanbot_input { border-top:1px solid #eef4fb; display:flex; gap:8px; padding:8px; background:#fbfdff;}
#finanbot_input input { flex:1; padding:8px 10px; border-radius:8px; border:1px solid #dfeefc; outline:none;}
#finanbot_input button { padding:8px 10px; border-radius:8px; background:#164b88; color:white; border:none; cursor:pointer; font-weight:600;}

/* peque√±o badge sugerente */
#finanbot_toggle { position: fixed; right: 22px; bottom: 18px; width:56px; height:56px; border-radius:50%; background:#164b88; color:white; display:flex; align-items:center; justify-content:center; font-weight:700; z-index:9998; cursor:pointer; box-shadow: 0 8px 22px rgba(22,75,136,0.25);}
</style>
"""

# ---------------------------
# === HTML (PIDO: no tocar formato principal) ===
# ---------------------------
html = """
<div class="simucalc-container">
  <h1>SIMULCREDIT</h1>
  <div class="brand-line"></div>

  <label for="monto">Monto del cr√©dito (USD):</label>
  <input type="number" id="monto" placeholder="Ej. 15000" />

  <label for="tasa">Tasa de inter√©s anual (%):</label>
  <input type="number" id="tasa" placeholder="Ej. 7.5" />

  <label for="plazo">Plazo (meses):</label>
  <input type="number" id="plazo" placeholder="Ej. 36" />

  <button onclick="google.colab.kernel.invokeFunction('calc_credit_backend',
    [parseFloat(document.getElementById('monto').value) || 0,
     parseFloat(document.getElementById('tasa').value) || 0,
     parseInt(document.getElementById('plazo').value) || 0], {})">
     Calcular cr√©dito
  </button>

  <div style="margin-top:12px;">
    <button class='download-btn' onclick="google.colab.kernel.invokeFunction('abrir_tasas_inicial', [], {})">üè¶ Integrar tasas internacionales</button>
    <button class='download-btn' onclick="google.colab.kernel.invokeFunction('abrir_componente_educativo_inicial', [], {})">üìö Componente educativo</button>
  </div>
</div>

<div id="resultado"></div>

<!-- Finanbot widget (inicialmente hidden; se mostrar√° s√≥lo en segunda p√°gina) -->
<div id="finanbot_widget">
  <div id="finanbot_header">
    <div>ü§ñ Finanbot</div>
    <div style="font-size:0.9rem; opacity:0.9;"><span id="finanbot_status">offline</span></div>
  </div>
  <div id="finanbot_messages"></div>
  <div id="finanbot_input">
    <input id="finanbot_text" placeholder="Pregunta sobre tu cr√©dito, pagos extra, plazo, tasas..." />
    <button onclick="enviarPreguntaFinanbot()">Enviar</button>
  </div>
</div>

<div id="finanbot_toggle" title="Abrir Finanbot" style="display:none;">üí¨</div>

<footer>¬© 2025 Simulcredit | Plataforma de simulaci√≥n financiera</footer>
"""

display(HTML(style + html))

# ---------------------------
# === FUNCIONES FINANCIERAS ===
# ---------------------------
def generar_tabla_amortizacion(monto, tasa_anual, plazo_meses):
    if plazo_meses <= 0:
        return pd.DataFrame([], columns=["Mes", "Pago total", "Capital", "Inter√©s", "Saldo restante"])
    if tasa_anual <= 0:
        cuota = monto / plazo_meses
        saldo = monto
        datos = []
        for i in range(1, plazo_meses + 1):
            interes = 0.0
            capital = cuota
            saldo -= capital
            datos.append([i, round(cuota,2), round(capital,2), round(interes,2), round(max(saldo,0),2)])
        return pd.DataFrame(datos, columns=["Mes", "Pago total", "Capital", "Inter√©s", "Saldo restante"])
    tasa_mensual = tasa_anual / 12 / 100
    cuota = monto * (tasa_mensual * (1 + tasa_mensual)**plazo_meses) / ((1 + tasa_mensual)**plazo_meses - 1)
    saldo = monto
    datos = []
    for i in range(1, plazo_meses + 1):
        interes = saldo * tasa_mensual
        capital = cuota - interes
        saldo -= capital
        datos.append([i, round(cuota, 2), round(capital, 2), round(interes, 2), round(max(saldo,0), 2)])
    return pd.DataFrame(datos, columns=["Mes", "Pago total", "Capital", "Inter√©s", "Saldo restante"])

# Estado global (segunda p√°gina)
ultima_tasa = None
ultimo_plazo = None
ultimo_monto = None
ultima_df = None

# ---------------------------
# === CALLBACK: calcular y mostrar segunda p√°gina ===
# ---------------------------
def calc_credit_backend(monto, tasa, plazo):
    global ultima_tasa, ultimo_plazo, ultimo_monto, ultima_df
    try:
        monto = float(monto)
        tasa = float(tasa)
        plazo = int(plazo)
    except:
        display(Javascript("""document.getElementById('resultado').innerHTML =
            '<b style="color:#b30000;">Por favor ingresa valores v√°lidos.</b>'"""))
        return
    if monto <= 0 or plazo <= 0 or tasa < 0:
        display(Javascript("""document.getElementById('resultado').innerHTML =
            '<b style="color:#b30000;">Por favor ingresa valores v√°lidos (monto>0, plazo>0, tasa‚â•0).</b>'"""))
        return

    df = generar_tabla_amortizacion(monto, tasa, plazo)
    ultima_df = df.copy()
    ultima_tasa, ultimo_plazo, ultimo_monto = tasa, plazo, monto

    # Guardar Excel (mismo comportamiento)
    nombre_archivo = "Tabla_Amortizacion_Simulcredit.xlsx"
    excel_buffer = io.BytesIO()
    df.to_excel(excel_buffer, index=False, engine='openpyxl')
    excel_buffer.seek(0)
    with open(nombre_archivo, "wb") as f:
        f.write(excel_buffer.read())

    # Resumen crediticio
    cuota = df['Pago total'].iloc[0] if not df.empty else 0
    total_pagado = df['Pago total'].sum()
    total_intereses = df['Inter√©s'].sum()
    porcentaje_intereses = (total_intereses / total_pagado * 100) if total_pagado>0 else 0

    resumen_html = f"""
      <div style="background:#eef6ff; padding:15px; border-radius:10px; margin-bottom:15px;">
        <b>Resumen crediticio:</b><br>
        Cuota mensual: <b>${cuota:,.2f}</b><br>
        Total pagado: <b>${total_pagado:,.2f}</b><br>
        Intereses totales: <b>${total_intereses:,.2f}</b>
      </div>
    """

    # Tabla EXACTA (sin tocar formato)
    def fmt(x): return f"{x:,.2f}"
    html_tabla = """
    <h3 style="color:#12355b; margin-top:0;">Tabla de Amortizaci√≥n</h3>
    <table class="table-styles">
      <thead>
        <tr>
          <th>Mes</th><th>Pago total (USD)</th><th>Capital (USD)</th>
          <th>Inter√©s (USD)</th><th>Saldo restante (USD)</th>
        </tr>
      </thead><tbody>
    """
    for _, row in df.iterrows():
        html_tabla += f"""
          <tr>
            <td style='text-align:left;'>{int(row['Mes'])}</td>
            <td>{fmt(row['Pago total'])}</td>
            <td>{fmt(row['Capital'])}</td>
            <td>{fmt(row['Inter√©s'])}</td>
            <td>{fmt(row['Saldo restante'])}</td>
          </tr>"""
    html_tabla += "</tbody></table>"

    # Botones en ORDEN: Dashboard, Finanbot (toggle), Simulador
    botones = """
      <div style="margin-top:12px;">
        <button class='download-btn' onclick="google.colab.kernel.invokeFunction('descargar_excel', [], {})">üì• Descargar Excel</button>
      </div>

      <div style="margin-top:10px;">
        <button class='download-btn' onclick="google.colab.kernel.invokeFunction('abrir_dashboard', [], {})">üìä Dashboard</button>
      </div>

      <div style="margin-top:8px;">
        <button class='download-btn' onclick="google.colab.kernel.invokeFunction('activar_finanbot_widget', [], {})">üí¨ Finanbot</button>
      </div>

      <div style="margin-top:8px;">
        <button class='download-btn' onclick="google.colab.kernel.invokeFunction('abrir_simulador_modal', [], {})">üîÅ Simulador de escenarios</button>
      </div>
    """

    # Append also the earlier buttons for "componente educativo" y "tasas" debajo (opcional)
    botones_inferiores = """
      <div style="margin-top:12px;">
        <button class='download-btn' onclick="google.colab.kernel.invokeFunction('abrir_componente_educativo_segunda', [], {})">üìö Componente educativo</button>
        <button class='download-btn' style="margin-left:10px;" onclick="google.colab.kernel.invokeFunction('abrir_tasas_segunda', [], {})">üè¶ Integrar tasas internacionales</button>
      </div>
    """

    display(Javascript(f"""
      document.getElementById('resultado').innerHTML = `{resumen_html + html_tabla + botones + botones_inferiores}`;
      // show Finanbot toggle only in second page (hidden earlier)
      document.getElementById('finanbot_toggle').style.display = 'flex';
    """))

# ---------------------------
# === DESCARGAR EXCEL ===
# ---------------------------
def descargar_excel():
    try:
        files.download("Tabla_Amortizacion_Simulcredit.xlsx")
    except Exception as e:
        display(Javascript(f"document.getElementById('resultado').innerHTML += '<div style=\"color:#b30000; margin-top:8px;\">Error al descargar: {str(e)}</div>';"))

# ---------------------------
# === DASHBOARD (profesional) ===
# ---------------------------
def abrir_dashboard():
    global ultima_df, ultimo_monto, ultimo_plazo, ultima_tasa
    if ultima_df is None:
        display(Javascript("document.getElementById('resultado').innerHTML += '<div style=\"color:#b30000; margin-top:8px;\">Calcula un cr√©dito primero.</div>';"))
        return
    df = ultima_df.copy()
    cuota = df['Pago total'].iloc[0]
    total_pagado = df['Pago total'].sum()
    total_intereses = df['Inter√©s'].sum()
    total_capital = df['Capital'].sum()

    # Gr√°ficos
    # 1) Saldo vs Mes
    plt.figure(figsize=(5,2.2))
    plt.plot(df['Mes'], df['Saldo restante'], marker='o', linewidth=2)
    plt.title('Saldo restante vs Mes')
    plt.xlabel('Mes')
    plt.ylabel('Saldo restante (USD)')
    plt.tight_layout()
    buf1 = io.BytesIO(); plt.savefig(buf1, format='png', dpi=120, bbox_inches='tight'); plt.close(); buf1.seek(0)
    b641 = base64.b64encode(buf1.read()).decode('utf-8')

    # 2) Pagos acumulados
    cum_cap = np.cumsum(df['Capital'])
    cum_int = np.cumsum(df['Inter√©s'])
    plt.figure(figsize=(5,2.2))
    ind = df['Mes']
    plt.bar(ind, cum_cap, label='Capital acumulado')
    plt.bar(ind, cum_int, bottom=cum_cap, label='Inter√©s acumulado')
    plt.xlabel('Mes'); plt.ylabel('USD acumulado'); plt.legend(); plt.tight_layout()
    buf2 = io.BytesIO(); plt.savefig(buf2, format='png', dpi=120, bbox_inches='tight'); plt.close(); buf2.seek(0)
    b642 = base64.b64encode(buf2.read()).decode('utf-8')

    # 3) Pie
    plt.figure(figsize=(3.5,3.5))
    plt.pie([total_capital, total_intereses], labels=['Capital','Inter√©s'], autopct='%1.1f%%', startangle=90)
    plt.axis('equal')
    buf3 = io.BytesIO(); plt.savefig(buf3, format='png', dpi=120, bbox_inches='tight'); plt.close(); buf3.seek(0)
    b643 = base64.b64encode(buf3.read()).decode('utf-8')

    dash_html = f"""
      <div style="margin-top:12px; padding:12px; border-radius:10px; background:#fff; border:1px solid #e8f1ff;">
        <h3 style="margin-top:0;color:#12355b;">üìä Dashboard profesional</h3>
        <div style="display:flex; gap:10px; flex-wrap:wrap;">
          <div style="flex:1; min-width:240px;">
            <b>KPI</b>
            <div style="margin-top:8px; padding:10px; background:#f7fbff; border-radius:8px;">
              <div><b>Monto:</b> ${ultimo_monto:,.2f}</div>
              <div><b>Plazo:</b> {ultimo_plazo} meses</div>
              <div><b>Cuota mensual:</b> ${cuota:,.2f}</div>
              <div><b>Total pagado:</b> ${total_pagado:,.2f}</div>
              <div><b>Intereses totales:</b> ${total_intereses:,.2f}</div>
            </div>
            <div style="margin-top:8px;"><b>Saldo restante (proyecci√≥n)</b>
              <div style="margin-top:8px;"><img src="data:image/png;base64,{b641}" style="width:100%; border-radius:8px;" /></div>
            </div>
          </div>

          <div style="width:220px;">
            <b>Composici√≥n de pago</b>
            <div style="margin-top:8px;"><img src="data:image/png;base64,{b643}" style="width:100%; border-radius:8px;" /></div>
            <div style="margin-top:8px;"><b>Acumulado</b><div><img src="data:image/png;base64,{b642}" style="width:100%; border-radius:8px;" /></div></div>
          </div>
        </div>
        <p style="color:#4b6b8d; margin-top:10px;"><small>Gr√°ficos generados a partir de la tabla actual.</small></p>
      </div>
    """
    display(Javascript(f"document.getElementById('resultado').innerHTML += `{dash_html}`;"))

# ---------------------------
# === FINANBOT: widget funcional (local heuristics + optional OpenAI) ===
# ---------------------------

# Helper: try to use OpenAI if key present.
USE_OPENAI = False
OPENAI_API_KEY = os.environ.get('OPENAI_API_KEY', None)
if OPENAI_API_KEY:
    USE_OPENAI = True

# Local lightweight responder (heuristics & calculations). Keeps scope limited to amortization domain.
def local_finanbot_response(question):
    global ultima_df, ultimo_monto, ultima_tasa, ultimo_plazo
    q = question.lower().strip()

    # If asks about amortization definition
    if re.search(r'\b(amortiz|qu√© es amortiz|qu√© es la amortiz|amortizaci√≥n)\b', q):
        return "La amortizaci√≥n es el proceso de pagar una deuda poco a poco mediante cuotas. Cada cuota contiene capital (lo que reduces) y inter√©s (el costo por el pr√©stamo)."

    # If asks about pagar extra
    m = re.search(r'pag(ar|o).*(extra|adicional|m√°s)\s*(\d+\.?\d*)', q)
    if m:
        try:
            extra = float(m.group(3))
        except:
            extra = round(ultimo_monto * 0.05, 2) if ultimo_monto else 0.0
        if ultima_df is None:
            return "Genera la tabla primero para que calcule exactamente el efecto del pago extra."
        # simulate with simple loop
        tasa = ultimo_tasa; monto = ultimo_monto
        tasa_m = tasa/12/100
        saldo = monto
        cuota = ultima_df['Pago total'].iloc[0]
        pagos = 0
        intereses_acum = 0.0
        while saldo > 0.001 and pagos < 2000:
            interes = saldo * tasa_m
            pago_total = cuota + extra
            capital = max(0, pago_total - interes)
            saldo -= capital
            intereses_acum += interes
            pagos += 1
            if saldo <= 0:
                break
        ahorro = ultima_df['Inter√©s'].sum() - intereses_acum
        return f"Si a√±ades ${extra:,.2f} extra al mes, pagar√≠as aproximadamente en {pagos} meses y ahorrar√≠as ~${ahorro:,.2f} en intereses (estimaci√≥n)."

    # If asks about refinancing
    if re.search(r'(refinanc|refinar|renegoc|cambiar.*tasa|tasa.*baj)', q):
        if ultimo_tasa is None:
            return "Genera la tabla primero para que pueda estimar el beneficio de refinanciar."
        objetivo = max(0.0, ultimo_tasa * 0.85)
        return f"Una meta razonable es intentar negociar por debajo de {objetivo:.2f}% anual. Cada punto porcentual menos reduce significativamente intereses."

    # If asks about plazo vs cuota
    if re.search(r'\b(plazo|cuota|meses|mensual)\b', q):
        return "Reducir el plazo baja los intereses totales pero aumenta la cuota mensual. Aumentar el plazo baja la cuota mensual pero aumenta intereses en total. Elige seg√∫n flujo de caja."

    # If asks for total interests
    if re.search(r'(interes(es)?|costo total|total a pagar)', q):
        if ultima_df is None:
            return "Genera la tabla primero para ver los intereses totales."
        total_intereses = ultima_df['Inter√©s'].sum()
        total_pagado = ultima_df['Pago total'].sum()
        return f"Intereses totales: ${total_intereses:,.2f}. Total pagado (capital+intereses): ${total_pagado:,.2f}."

    # If numeric question "cuanto ahorro"
    m2 = re.search(r'ahorr.*\s*(\d+\.?\d*)', q)
    if 'ahorro' in q and ultima_df is not None and m2:
        # assume asking saving from extra
        extra = float(m2.group(1))
        # reuse earlier logic
        tasa = ultimo_tasa; monto = ultimo_monto
        tasa_m = tasa/12/100
        saldo = monto
        cuota = ultima_df['Pago total'].iloc[0]
        pagos=0; intereses_acum=0.0
        while saldo>0.001 and pagos<2000:
            interes = saldo*tasa_m
            pago_total = cuota + extra
            capital = max(0, pago_total - interes)
            saldo -= capital
            intereses_acum += interes
            pagos+=1
            if saldo<=0: break
        ahorro = ultima_df['Inter√©s'].sum() - intereses_acum
        return f"Con pago extra ${extra:,.2f} al mes, estimas ahorrar ~${ahorro:,.2f} y terminar en {pagos} meses."

    # fallback short domain-limited answer
    return ("Disculpa, puedo ayudar con dudas sobre esta tabla: amortizaci√≥n, pagos extra, plazo, tasa y refinanciamiento. "
            "Prueba: '¬øCu√°nto ahorro si pago extra 100?' o '¬øconviene refinanciar a 6.5%?'")

# Optional OpenAI usage (if user provides OPENAI_API_KEY). Use carefully and only for finance domain.
def openai_finanbot_response(question):
    try:
        # Lazy import here to avoid hard dependency
        from openai import OpenAI
    except Exception as e:
        return None
    key = os.environ.get('OPENAI_API_KEY', None)
    if not key:
        return None
    # If openai available, call a short completions / chat
    try:
        client = OpenAI(api_key=key)
        prompt = (
            "Eres un asistente financiero conciso que solo responde preguntas sobre amortizaci√≥n, pagos extra, refinanciamiento, plazos y tasas.\n"
            f"Contexto: monto={ultimo_monto}, tasa={ultimo_tasa}, plazo={ultimo_plazo} meses.\n"
            f"Pregunta: {question}\n"
            "Responde en espa√±ol, breve y con n√∫meros si es necesario."
        )
        resp = client.responses.create(model="gpt-4o-mini", input=prompt, max_tokens=250)
        out = resp.output_text if hasattr(resp, 'output_text') else str(resp)
        return out
    except Exception as e:
        return None

# JS bridge: activar widget y enviar preguntas al backend
def activar_finanbot_widget():
    # show toggle and widget
    display(Javascript("""
      document.getElementById('finanbot_widget').style.display = 'block';
      document.getElementById('finanbot_toggle').style.display = 'none';
      document.getElementById('finanbot_status').innerText = 'listening';
      // ensure messages area shows a welcome
      const m = document.getElementById('finanbot_messages');
      m.innerHTML = "<div style='margin-bottom:8px;'><b>Finanbot:</b> Hola ‚Äî soy Finanbot. Pregunta sobre la tabla (pagos extra, amortizaci√≥n, refinanciamiento).</div>";
      m.scrollTop = m.scrollHeight;
    """))

# Function called from JS to process question (registered as callback)
def procesar_pregunta_finanbot(texto):
    texto = str(texto)
    # If OpenAI is configured and permitted, try it first for a more natural answer
    if USE_OPENAI:
        try:
            out = openai_finanbot_response(texto)
            if out:
                return out
        except Exception:
            pass
    # fallback to local response
    return local_finanbot_response(texto)

# JS helper: define enviarPreguntaFinanbot function in the notebook frontend
display(Javascript("""
window.enviarPreguntaFinanbot = function(){
  const txt = document.getElementById('finanbot_text').value || '';
  if(!txt.trim()) return;
  const area = document.getElementById('finanbot_messages');
  area.innerHTML += "<div style='margin-bottom:8px;'><b>T√∫:</b> " + txt.replace(/</g,'&lt;') + "</div>";
  area.scrollTop = area.scrollHeight;
  document.getElementById('finanbot_text').value = '';
  // call python backend
  google.colab.kernel.invokeFunction('procesar_pregunta_finanbot', [txt], {}).then((resp) => {
    // resp is text
    area.innerHTML += "<div style='margin-bottom:8px; color:#0a2540;'><b>Finanbot:</b> " + resp.replace(/</g,'&lt;') + "</div>";
    area.scrollTop = area.scrollHeight;
  });
};
"""))

# Register callback
output.register_callback('procesar_pregunta_finanbot', procesar_pregunta_finanbot)
output.register_callback('activar_finanbot_widget', activar_finanbot_widget)

# Also register a callback to close/hide Finanbot and show small toggle button
display(Javascript("""
document.getElementById('finanbot_toggle').onclick = function(){
  // show widget
  document.getElementById('finanbot_widget').style.display = 'block';
  this.style.display = 'none';
};
"""))

# ---------------------------
# === SIMULADOR (modal-like insert) ===
# ---------------------------
def abrir_simulador_modal():
    # inject a nice modal-like panel into resultado (keeps table visible)
    if ultima_df is None:
        display(Javascript("document.getElementById('resultado').innerHTML += '<div style=color:#b30000; margin-top:8px;>Calcula un cr√©dito primero para usar el simulador.</div>';"))
        return
    js = r"""
    const cont = document.getElementById('resultado');
    // create panel
    cont.innerHTML += `
      <div id="sim_modal" style="margin-top:12px; padding:12px; border-radius:10px; background:linear-gradient(180deg,#fff,#f7fcff); border:1px solid #e6f4ff;">
        <h3 style="margin-top:0; color:#12355b;">üîÅ Simulador de escenarios</h3>
        <div>
          <label>Tasa alternativa (%)</label>
          <input id="sim_tasa" type="number" placeholder="Ej. 6.5" />
          <label style="margin-top:8px;">Plazo alternativo (meses)</label>
          <input id="sim_plazo" type="number" placeholder="Ej. 24" />
        </div>
        <div style="margin-top:10px;">
          <button class="download-btn" onclick="google.colab.kernel.invokeFunction('procesar_simulador', [parseFloat(document.getElementById('sim_tasa').value)||0, parseInt(document.getElementById('sim_plazo').value)||0], {})">Comparar</button>
          <button class="download-btn" onclick="document.getElementById('sim_modal').remove()" style="margin-left:8px;">Cerrar</button>
        </div>
      </div>
    `;
    window.scrollTo(0, document.getElementById('sim_modal').offsetTop - 20);
    """
    display(Javascript(js))

def procesar_simulador(tasa_alt, plazo_alt):
    global ultima_df, ultimo_monto, ultima_tasa, ultimo_plazo
    if ultima_df is None:
        display(Javascript("document.getElementById('resultado').innerHTML += '<div style=color:#b30000; margin-top:8px;>Calcula un cr√©dito primero.</div>';"))
        return
    try:
        tasa_alt = float(tasa_alt); plazo_alt = int(plazo_alt)
    except:
        display(Javascript("document.getElementById('resultado').innerHTML += '<div style=color:#b30000; margin-top:8px;>Ingresa valores v√°lidos.</div>';"))
        return
    if tasa_alt<=0 or plazo_alt<=0:
        display(Javascript("document.getElementById('resultado').innerHTML += '<div style=color:#b30000; margin-top:8px;>Valores mayores a 0.</div>';"))
        return
    monto = ultimo_monto
    df_base = ultima_df
    df_alt = generar_tabla_amortizacion(monto, tasa_alt, plazo_alt)
    total_inter_base = df_base['Inter√©s'].sum()
    total_inter_alt = df_alt['Inter√©s'].sum()
    ahorro = total_inter_base - total_inter_alt
    # pretty verdict
    if ahorro > 0:
        sugerido = f"Te conviene el escenario alternativo: ahorras ${ahorro:,.2f} en intereses."
    elif ahorro < 0:
        sugerido = f"No conviene el escenario alternativo: pagar√≠as ${-ahorro:,.2f} m√°s en intereses."
    else:
        sugerido = "Ambos escenarios dan resultados similares."

    # grafico
    plt.figure(figsize=(4,2.2))
    bars = [total_inter_base, total_inter_alt]
    labels = ['Actual','Alternativa']
    x = np.arange(len(labels))
    plt.bar(x, bars)
    plt.xticks(x, labels)
    plt.ylabel('Intereses totales (USD)')
    plt.tight_layout()
    buf = io.BytesIO(); plt.savefig(buf, format='png', dpi=120, bbox_inches='tight'); plt.close(); buf.seek(0)
    b64 = base64.b64encode(buf.read()).decode('utf-8')

    html = f"""
      <div style="margin-top:12px; padding:12px; border-radius:10px; background:#f0f9ff; border:1px solid #dff0ff;">
        <h4 style="margin-top:0; color:#12355b;">Resultado</h4>
        <p><b>Intereses actuales:</b> ${total_inter_base:,.2f}</p>
        <p><b>Intereses alternativa:</b> ${total_inter_alt:,.2f}</p>
        <p><b>Ahorro estimado:</b> ${ahorro:,.2f}</p>
        <p style="font-weight:700;">{sugerido}</p>
        <p>Si quieres profundizar, pregunta al Finanbot en la esquina inferior derecha.</p>
        <div style="margin-top:8px;"><img src="data:image/png;base64,{b64}" style="width:100%; border-radius:8px;" /></div>
      </div>
    """
    display(Javascript(f"document.getElementById('resultado').innerHTML += `{html}`;"))

# ---------------------------
# === COMPONENTE EDUCATIVO (ANTES de generar tabla) ===
# ---------------------------
def abrir_componente_educativo_inicial():
    # This opens a "page" explaining basics ‚Äî before table generation
    educ_html = r"""
    <div style="padding:12px; border-radius:12px; background:linear-gradient(180deg,#fff,#f8fbff);">
      <h2 style="color:#12355b; margin-top:0;">üìö Componente educativo</h2>
      <div style="background:#fff; padding:10px; border-radius:10px; border:1px solid #eef4fb;">
        <h3>¬øQu√© es un cr√©dito? (explicado f√°cil)</h3>
        <p>Es pedir dinero para comprar algo ahora y devolverlo poco a poco. Piensa: pides una bici, y pagas mensualmente hasta completarla.</p>
        <svg width="100%" height="90" viewBox="0 0 200 90" xmlns="http://www.w3.org/2000/svg">
          <rect x="8" y="30" rx="10" ry="10" width="60" height="30" fill="#f3f8fc" stroke="#a7c4e0"/>
          <text x="12" y="50" fill="#12355b" font-size="12">Bici ahora ‚Üí</text>
        </svg>
        <hr>
        <h4>Conceptos clave</h4>
        <ul>
          <li><b>Tasa de inter√©s:</b> precio por pedir prestado.</li>
          <li><b>Plazo:</b> cu√°ntos meses tardas en pagar.</li>
          <li><b>Cuota:</b> lo que pagas cada mes (capital + inter√©s).</li>
        </ul>
        <div style="margin-top:8px;">
          <button class="download-btn" onclick="google.colab.kernel.invokeFunction('volver_inicio', [], {})">Volver</button>
        </div>
      </div>
    </div>
    """
    display(Javascript(f"document.getElementById('resultado').innerHTML = `{educ_html}`;"))

def volver_inicio():
    # re-render the original main container (we will simply clear resultado)
    display(Javascript("document.getElementById('resultado').innerHTML = '';"))

# ---------------------------
# === TASAS INTERNACIONALES (ANTES: initial menu) ===
# ---------------------------

# We'll try to fetch some central bank rates; many official APIs require keys.
# We'll try simple scraping from known pages as fallback.
def fetch_tasa_mexico():
    # Banxico: best to use their API (requires token). Fallback: return None to indicate unauthenticated.
    return None

def fetch_tasa_colombia():
    # Banco de la Republica: scraping possible - we fallback to None
    return None

def fetch_tasa_eeuu():
    # Try to get federal funds effective rate from St. Louis FRED via public CSV (may require API). fallback None
    return None

def fetch_tasa_brasil():
    return None

def fetch_tasa_espana():
    return None

# To keep behavior deterministic: if fetch functions return None, we show message and offer static fallback.
tasas_fallback = {
    "mexico": {"name":"Banco de M√©xico", "tasa":"(necesita API o scraping)"},
    "colombia": {"name":"Banco de la Rep√∫blica (Colombia)", "tasa":"(necesita API o scraping)"},
    "eeuu": {"name":"Federal Reserve (EE.UU.)", "tasa":"(necesita API o scraping)"},
    "brasil": {"name":"Banco Central do Brasil", "tasa":"(necesita API o scraping)"},
    "espana": {"name":"Banco de Espa√±a / BCE", "tasa":"(necesita API o scraping)"},
}

def abrir_tasas_inicial():
    # show initial menu before generating table
    menu_html = """
      <div style="margin-top:12px; padding:12px; border-radius:10px; background:#fff; border:1px solid #e8f1ff;">
        <h3 style="margin-top:0;color:#12355b;">üè¶ Tasas bancarias internacionales (actuales)</h3>
        <p style="color:#4b6b8d;">Intentando obtener las tasas reales... si tu entorno no permite llamadas externas, ver√°s una nota y valores gu√≠a.</p>
        <div style="display:flex; gap:8px; flex-wrap:wrap;">
          <button class="download-btn" onclick="google.colab.kernel.invokeFunction('mostrar_tasa_key_init', ['mexico'], {})">Banco de M√©xico</button>
          <button class="download-btn" onclick="google.colab.kernel.invokeFunction('mostrar_tasa_key_init', ['colombia'], {})">Colombia</button>
          <button class="download-btn" onclick="google.colab.kernel.invokeFunction('mostrar_tasa_key_init', ['eeuu'], {})">EE.UU.</button>
          <button class="download-btn" onclick="google.colab.kernel.invokeFunction('mostrar_tasa_key_init', ['brasil'], {})">Brasil</button>
          <button class="download-btn" onclick="google.colab.kernel.invokeFunction('mostrar_tasa_key_init', ['espana'], {})">Espa√±a</button>
        </div>
      </div>
    """
    display(Javascript(f"document.getElementById('resultado').innerHTML = `{menu_html}`;"))

def mostrar_tasa_key_init(key):
    # Try to fetch real - we implemented fetch_* above and left them as None fallback for now.
    mapping = {
        "mexico": fetch_tasa_mexico,
        "colombia": fetch_tasa_colombia,
        "eeuu": fetch_tasa_eeuu,
        "brasil": fetch_tasa_brasil,
        "espana": fetch_tasa_espana,
    }
    fn = mapping.get(key)
    tasa = None
    if fn:
        try:
            tasa = fn()
        except:
            tasa = None
    if tasa is None:
        d = tasas_fallback.get(key)
        html = f"""
          <div style="margin-top:12px; padding:12px; border-radius:10px; background:#f8fff6; border:1px solid #e8f6ea;">
            <h4 style="margin-top:0;color:#12355b;">{d['name']}</h4>
            <p><b>Tasa de referencia (actual):</b> {d['tasa']}</p>
            <p style="color:#4b6b8d;"><small>Nota: en este entorno Colab puede que no se pueda acceder a APIs oficiales. Si quieres, puedo a√±adir c√≥digo espec√≠fico para Banxico, FRED o BanRep (requieren tokens o end-points). ¬øDeseas que lo a√±ada?</small></p>
          </div>
        """
        display(Javascript(f"document.getElementById('resultado').innerHTML += `{html}`;"))
    else:
        # if tasa was retrieved
        html = f"""
          <div style="margin-top:12px; padding:12px; border-radius:10px; background:#f8fff6; border:1px solid #e8f6ea;">
            <h4 style="margin-top:0;color:#12355b;">Fuente</h4>
            <p><b>Tasa de referencia (actual):</b> {tasa}</p>
          </div>
        """
        display(Javascript(f"document.getElementById('resultado').innerHTML += `{html}`;"))

# Second-page versions (same behavior)
def abrir_tasas_segunda():
    abrir_tasas_inicial()

# ---------------------------
# === COMPONENTE EDUCATIVO (SEGUNDA P√ÅGINA) ===
# ---------------------------
def abrir_componente_educativo_segunda():
    # More completo y cute (reemplaza #resultado)
    educ_html = r"""
      <div style="padding:12px; border-radius:12px; background:linear-gradient(180deg,#fff,#f8fbff);">
        <h2 style="color:#12355b; margin-top:0;">üìö Componente educativo ‚Äî explicado f√°cil</h2>
        <div style="background:#fff; padding:10px; border-radius:10px; border:1px solid #eef4fb;">
          <h3>¬øQu√© es un cr√©dito?</h3>
          <p>Es pedir dinero ahora y devolverlo poco a poco. Imagina que compras una bici y pagas mes a mes hasta que ya no debes nada.</p>
          <svg width="100%" height="80" viewBox="0 0 200 80" xmlns="http://www.w3.org/2000/svg">
            <circle cx="30" cy="50" r="12" fill="#ffe8e8" stroke="#ffc0c0"/>
            <rect x="60" y="28" rx="8" ry="8" width="110" height="36" fill="#eef6ff" stroke="#d7e7ff"/>
            <text x="66" y="50" fill="#12355b" font-size="12">Pides dinero ‚Üí Pagas en cuotas</text>
          </svg>
          <hr>
          <h4>M√°s conceptos</h4>
          <ul>
            <li><b>Tasa de inter√©s:</b> cu√°nto te cuesta pedir prestado.</li>
            <li><b>Plazo:</b> n√∫mero de meses para pagar.</li>
            <li><b>Cuota:</b> pago mensual (capital + inter√©s).</li>
          </ul>
          <h4>Tips f√°ciles</h4>
          <ol>
            <li>Paga extra si puedes: reduce tiempo y intereses.</li>
            <li>No uses m√°s del 30-35% de tus ingresos en deudas.</li>
            <li>Compara tasas antes de firmar un cr√©dito.</li>
          </ol>
          <div style="margin-top:8px;">
            <button class="download-btn" onclick="google.colab.kernel.invokeFunction('volver_a_tabla', [], {})">Volver a la tabla</button>
          </div>
        </div>
      </div>
    """
    display(Javascript(f"document.getElementById('resultado').innerHTML = `{educ_html}`;"))

# Volver a tabla
def volver_a_tabla():
    if ultima_df is not None:
        calc_credit_backend(ultimo_monto, ultima_tasa, ultimo_plazo)
    else:
        display(Javascript("document.getElementById('resultado').innerHTML = '<b>Calcula un cr√©dito primero.</b>';"))

# ---------------------------
# === REGISTRO DE CALLBACKS ===
# ---------------------------
output.register_callback('calc_credit_backend', calc_credit_backend)
output.register_callback('descargar_excel', descargar_excel)
output.register_callback('abrir_dashboard', abrir_dashboard)
output.register_callback('activar_finanbot_widget', activar_finanbot_widget)
output.register_callback('procesar_pregunta_finanbot', procesar_pregunta_finanbot)
output.register_callback('abrir_simulador_modal', abrir_simulador_modal)
output.register_callback('procesar_simulador', procesar_simulador)
output.register_callback('abrir_componente_educativo_inicial', abrir_componente_educativo_inicial)
output.register_callback('abrir_tasas_inicial', abrir_tasas_inicial)
output.register_callback('mostrar_tasa_key_init', mostrar_tasa_key_init)
output.register_callback('abrir_componente_educativo_segunda', abrir_componente_educativo_segunda)
output.register_callback('abrir_tasas_segunda', abrir_tasas_segunda)
output.register_callback('volver_inicio', volver_inicio)
output.register_callback('volver_a_tabla', volver_a_tabla)

print("Listo ‚Äî pega valores y prueba: 1) usa 'Integrar tasas internacionales' y 'Componente educativo' antes de calcular; 2) calcula para ver Dashboard ‚Üí Finanbot (corner widget) ‚Üí Simulador.")

```
